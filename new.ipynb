```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

titanic = pd.read_csv('train.csv')

titanic.head(3)
```

```python
titanic_col_num = titanic.shape[1]
print(f"Column numbers of train_file: {titanic_col_num}")
```

```python
titanic_row_num = titanic.shape[0]
print(f"Row numbers of train_file: {titanic_row_num}")
```

```python
titanic_age_max = max(titanic['Age'])
print(f"Oldest Passenger: {titanic_age_max}")
```

```python
titanic_age_min = min(titanic['Age'])
print(f"Younger Passenger: {titanic_age_min}")
```

```python
titanic_average_age = np.mean(titanic['Age'])
print(f"Average Age of Titanic: {round(titanic_average_age, 2)}")
```

```python
titanic.dropna(axis=0, inplace=True)

titanic
```

```python
titanic.info()
```

    The following example converts "Pclass" into data type float64.

```python
titanic['Pclass'] = titanic['Pclass'].astype(float)
titanic.info()
```

```python
titanic.describe()
```

```python
titanic_sortAge = titanic.sort_values(by='Age')
titanic_sortAge.plot(x='Age', y='Survived', kind='line')
plt.ylim(ymin=0)
plt.xlim(xmin=0)

plt.show()
```

```python
titanic_sortAge = titanic[titanic['Sex'] == 'male'].sort_values(by='Age')
titanic_sortAge.plot(x='Age', y='Survived', kind='line')
plt.ylim(ymin=0)
plt.xlim(xmin=0)

plt.show()
```

# kendi oluşturduğum data için çalışmalar

```python
my_data = {
    'İsim': ['Emre', 'İrem', 'Süheyla', 'Çukurlu', 'Çakır', 'Osman'],
    'Yaş': [10, 20, 30, 40, 50, 60],
    'Kilo': [50, 80, 110, 140, 170, 200]
}

df = pd.DataFrame(my_data)
df
```

```python
df['Yaş'].loc[0]
```

```python
df.plot(x='Yaş', y='Kilo', kind='line')
plt.ylim(ymin=0)
plt.xlim(xmin=0)

plt.show()
```

```python
def slope(x1, y1, x2, y2):
    s = (y2-y1)/(x2-x1)
    return s
```

```python
slope(df['Yaş'].loc[0],df['Kilo'].loc[0],df['Yaş'].loc[1],df['Kilo'].loc[1] )
```

OR

```python
x = df['Yaş']
y = df['Kilo']
slope_intercept = np.polyfit(x, y, 1)
slope_intercept
```

```python
def my_function(x):
    return round(slope_intercept[0])*x + round(slope_intercept[1])
```

```python
my_function(60)
```
