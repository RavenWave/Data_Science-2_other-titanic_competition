```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from module_emre import module_emre as mer
```

### NOT: SibSp ve Parch kolonlarından en az birinde bile 1 yazıyorsa, alone=1; her ikisi de sıfır ise alone=0 olacak yeşkilde "alone" kolonu ekleyelim.

```python
df = pd.read_csv("SOURCES/titanic_train.csv")
df.head()
```

```python
df["Alone"] = df["SibSp"] + df["Parch"]
df.head()
```

```python
df["Alone"].value_counts()
```

```python
df_alone = df["Alone"]
for i in range(len(df_alone)):
    if df_alone[i] > 0:
        df.loc[i, "Alone"] = 1
    # print(i)
df_alone = df["Alone"]
df["Alone"].value_counts()
```

```python
df.drop(["Cabin", "Ticket", "Name"] , axis=1, inplace=True)
df["Sex"].replace({"male": 0, "female": 1}, inplace=True)
df.head()
```

```python
df.info()
```

```python
mean_ages = df["Age"].median()
df.fillna(mean_ages, inplace=True)
df.info()
```

```python
df_ages = df[["Age"]]
list_ages = list(df_ages["Age"])

df_pclass = df[["Pclass"]]
list_pclass = list(df_pclass["Pclass"])

df_sex = df[["Sex"]]
list_sex = list(df_sex["Sex"])

df_alone = df[["Alone"]]
list_alone = list(df_alone["Alone"])

df_survived = df[["Survived"]]
list_survived = list(df_survived["Survived"])
```

```python
zipped_data = list(zip(list_pclass, list_ages, list_sex, list_alone))

my_list = []
for element in zipped_data:
    my_list.append(list(element))
    
X = np.array(my_list)
Y = np.array(list_survived)
```

```python
# from mpl_toolkits.mplot3d import Axes3D

# fig = plt.figure()
# ax = fig.add_subplot(111, projection='3d')

# ax.scatter(X[:, 0], X[:, 1], X[:, 2], X[:, 3], c=Y, cmap='viridis')
# ax.set_xlabel('pclass')
# ax.set_ylabel('age')
# ax.set_zlabel('fare')
# ax.set_wlabel('4. Özellik')

# plt.show()
```

```python
clf = GaussianNB()
clf.fit(X, Y)
```

```python
df_test = pd.read_csv("SOURCES/titanic_test.csv")
df.head()
```

```python
df_test.drop(["Cabin", "Ticket", "Name"] , axis=1, inplace=True)

df_test["Alone"] = df_test["SibSp"] + df_test["Parch"]
df_alone = df_test["Alone"]
for i in range(len(df_alone)):
    if df_alone[i] > 0:
        df_test.loc[i, "Alone"] = 1
    # print(i)
df_alone = df["Alone"]

df_test["Sex"].replace({"male": 0, "female": 1}, inplace=True)
mean_ages = df_test["Age"].median()
df_test.fillna(mean_ages, inplace=True)
df_test.info()
```

```python
df_agesT = df_test[["Age"]]
list_agesT = list(df_agesT["Age"])

df_pclassT = df_test[["Pclass"]]
list_pclassT = list(df_pclassT["Pclass"])

df_sexT = df_test[["Sex"]]
list_sexT = list(df_sexT["Sex"])

df_aloneT = df_test[["Alone"]]
list_aloneT = list(df_aloneT["Alone"])

zipped_data = list(zip(list_pclassT, list_agesT, list_sexT, list_aloneT))

my_list = []
for element in zipped_data:
    my_list.append(list(element))
    
X_test = np.array(my_list)
```

```python
result_train = clf.predict(X_test)
```
