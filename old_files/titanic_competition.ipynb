### third class erkek yolcular

```python
third_class = df_train[df_train['Pclass'] == 3]
third_class_male = third_class[third_class['Sex'] == 'male']
third_class_male_liveN = len(third_class_male[third_class_male['Survived'] == 1])
third_class_male_diedN = len(third_class_male[third_class_male['Survived'] == 0])
third_class_male_totalN = third_class_male_liveN + third_class_male_diedN

print(f"üçüncü sınıf erkek yolculardan ölen    sayısı --> {third_class_male_diedN}")
print(f"üçüncü sınıf erkek yolculardan ölmeyen sayısı --> {third_class_male_liveN}")
print(f"üçüncü sınıf erkek yolculardan ölen oranı     --> {round((100 * (third_class_male_liveN / third_class_male_totalN)))}%")
```

```python
third_class_maleQ = third_class_male[third_class_male['Embarked'] == "Q"]
third_class_maleQ.sort_values(by='Survived', ascending=False)
third_class_maleQ_died = len(third_class_maleQ[third_class_maleQ['Survived'] == 0])
third_class_maleQ_live = len(third_class_maleQ[third_class_maleQ['Survived'] == 1])
third_class_maleQ_total = third_class_maleQ_died + third_class_maleQ_live
x = str(round((100 * (third_class_maleQ_live/third_class_maleQ_total)))) + "%"

print(f"üçüncü sınıf bileti olan erkeklerin Q şehrinden gelenlerinin yaşama şansı: {x}")
print("\n⬆⬆ Bu sebepten dolayı, bu kategorideki kişilerin öleceğini varsayacağım. ")
```

```python
indexler = third_class_maleQ.index
for x in indexler:
    df_train = df_train.drop(index=x)
```
