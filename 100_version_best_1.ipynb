```python
import pandas as pd
import numpy as np
```

```python
df0 = pd.read_csv("SOURCES/titanic_train.csv")
df = df0.copy()
```

```python
df
```

```python
df["Alone"] = df["SibSp"] + df["Parch"]
```

```python
for i in range(len(df)):
    if df.iloc[i, -1] > 0:
        df.iloc[i, -1] = 1
    else:
        df.iloc[i, -1] = 0
```

```python
df
```

```python
df.drop(columns=["Name", "SibSp", "Parch", "Cabin", "Ticket"], inplace=True)
```

```python
df["Sex"].replace({"male":1, "female":2}, inplace=True)
df["Embarked"].replace({"S":1, "C":2, "Q":3}, inplace=True)
df["Age"].replace({np.nan:df["Age"].median()}, inplace=True)
df["Embarked"].replace({np.nan:"1"}, inplace=True)              
```

```python
df.head()
```

```python
df.info()
```

```python
df.corr()
```

```python
X = df.iloc[:, [2, 3, 4, 5, 7]].values
y = df.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=3)
clf.fit(X_train, y_train)
```

```python
print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```
